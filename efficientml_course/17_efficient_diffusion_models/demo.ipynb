```python
!pip install torch -q
import torch

# Diffusion Sampling Speed Comparison
def diffusion_time_estimate(steps, time_per_step=0.05):
    """Estimate generation time based on sampling steps"""
    return steps * time_per_step

methods = {
    'DDPM': 1000,
    'DDIM': 50,
    'DPM-Solver': 20,
    'LCM': 4,
    'SDXL Turbo': 1,
}

print("Diffusion Sampling Methods (512x512 image, A100)")
print("=" * 50)
print(f"{'Method':<15} | {'Steps':>6} | {'Time':>8} | Speedup")
print("-" * 50)

base_time = diffusion_time_estimate(methods['DDPM'])
for name, steps in methods.items():
    time = diffusion_time_estimate(steps)
    speedup = base_time / time
    bar = "â–ˆ" * min(int(speedup / 10), 20)
    print(f"{name:<15} | {steps:>6} | {time:>6.2f}s | {speedup:>5.0f}x {bar}")

print("\nðŸŽ¯ 1-step generation is 1000x faster than original!")
```
